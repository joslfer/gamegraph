# Games on networks

The goal of this notebook is to explore how different network topologies affect the evolution of cooperation, and to find a particular network were cooperation survives and stabilizes. 

## Experiment population 
Population Parameters:

- 100 agents
- 80% initial cooperator percentage
- 100 generations

```python
number_of_agents = 100
N = 100

p_coop = 0.5
population = np.random.binomial(1, p_coop, size = number_of_agents).astype(bool)
```

## Erdős–Rényi Random Graph

Parameters: 

```python
p_edge = 0.08
G1 = nx.erdos_renyi_graph(number_of_agents, p_edge, seed = 67)

G = G1
pos = nx.spring_layout(G, seed=67)
```

```python
history, cooperatio = simulate_network(population, G, N)
```

```python
ani = animate_network(history, G, pos, filename = "../images/erdos_renyi.gif")
mpl.rcParams["animation.embed_limit"] = 50  # MegaBytes
HTML(ani.to_jshtml())
```

```python
plot_cooperation(cooperatio, "Erdős–Rényi Graph")
```

# Conclusion: Finding stable cooperation 

It is still unclear wether network topology affects the stability of cooperator populations. At first, I thought that higher degrees benefited defectors, but I could not find convincing evidence to support this claim. The most important parameter seems to be R, the reward payoff when two cooperators play together. When R = 3 every network collapsed into a population of defectors, and with R = 4.5 cooperators stabilized.
