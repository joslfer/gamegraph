# Network Topologies

## Normal network 

Some radom network without any specific characteristics.

```python

G = nx.Graph()
G.add_nodes_from([1,2,3,4,5])
G.add_edges_from([(1,2),(4,1),(3,2),(3,1)]) #accepts a tuple list


nx.draw(G, with_labels = True)
plt.show()
```

## Erdős–Rényi Random Graph


Each possible connection between two nodes forms independently with probabilty p. The expeted number of connections in the graphs is:

$$p \cdot \frac{n(n-1)}{2}$$

```python
G1 = nx.erdos_renyi_graph(10, 0.3, seed = 67)
nx.draw(G1,with_labels=True)
```

## Grid (Lattice)

A network arranged as a 2D grid, where each node is connected only to its immediate neighbours (up, down, left, right)

```python
G_lattice = nx.grid_2d_graph(3, 8)
nx.draw(G_lattice,with_labels=True)
```

## Watts–Strogatz Small-World Network

In a Watts-Strogatz small world network a ring structure is formed where each node is connected to its k nearest neighbours. There is a probability p that connection is rewired connecting the node with any random node in the ring. As a result shortcuts are introduced and the mean distance between nodes decreases a lot.

```python
G_strogatz = nx.watts_strogatz_graph(10,4,0.1, seed=67)
nx.draw(G_strogatz)
plt.show()
```

## Barabási–Albert Scale-Free Network

In a Barabási-Albert Scale-Free network each new node is connected to m other existent nodes. 

These new connections are established using preferential attachment, where the probabilty of connecting to node i is proportional to its degree: 


$$P(i)=\frac{k_i}{\sum_j k_j}.$$

Because of this, the first nodes are more likley to recieve connections forming hubs.

```python
G_hubs = nx.barabasi_albert_graph(10, 3, seed=67, initial_graph=None)
nx.draw(G_hubs)
```
